##Preprocessing

To prepare our model for training, we need to convert our text data  into features that our model will use to train and cast future predictions. We'll use two preprocessing steps:


1.   Count vectorization text
        converts text documents into matrix of tokens count. Each token represents a column in the matrix
2.   tf-idf
        converts cleaned text into numerical format.



#vectorization techniques
  Bag of words method: this model converts text by converting it into collection of words and their frequencies.
   

```python
#checking count vectorization
from sklearn.feature_extraction.text import CountVectorizer
corpus = [
    'This is the first document',
    'This document is the second document',
    'and this is the third one',
    'is this the first document'
]
#fit the data to countvectorizer function
vec = CountVectorizer().fit(corpus)
#print out the feature names of the fitted countvectorizer data
vec.get_feature_names_out()
```

```python
#transform the data into array
vec.transform(corpus).toarray()
```
